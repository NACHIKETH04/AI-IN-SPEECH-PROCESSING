### A1.

#### Use librosa.effects.trim() to remove the silence parts of speech from beginning and end of your recorded signal. Listen to the new signal and perceptually compare the audio with original.

```python
y, sr = librosa.load('AI_VOICE.wav')
librosa.display.waveshow(y)
plt.title('Speech Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

#### Plotting the Original speech signal

```python
import librosa
import matplotlib.pyplot as plt

# Load the audio file
audio, sample_rate = librosa.load('AI_VOICE.wav')

# Display the original speech signal 
plt.figure(figsize=(10, 4))
librosa.display.waveshow(audio, sr=sample_rate, color='black')
plt.title('Original Speech Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

#### Trim the silent parts from the beginning and end of the signal and plot the trimmed signal

```python
y_trimmed_audio,index= librosa.effects.trim(y)

# Display the original speech signal 
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y_trimmed_audio, sr=sample_rate, color='red')
plt.title('Trimmed Speech Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

#### Comparing the audio's of trimmed and Original audio signal

```python
import IPython.display as ipd

print("Trimmed Speech Signal")
ipd.Audio(data=y_trimmed_audio,rate=sample_rate)
```

```python
print("Original Speech Signal")
ipd.Audio(data=audio,rate=sample_rate)
```

### A2. 

Use librosa.effects.split() to split the recorded speech with detected silences. Play around with the top_db parameter and see the effects of split. Listen to the generated signals and  observe the split quality.

```python
import librosa
from IPython.display import Audio

# Load the audio file 
audio_data, sample_rate = librosa.load('AI_VOICE.wav')

# Split the audio into segments based on silence 
audio_segments = librosa.effects.split(audio_data, top_db=30)

num_segments = min(5, len(audio_segments))
end_idx_previous = 0

for i in range(num_segments):
    segment_bounds = audio_segments[i]
    start_idx, end_idx = segment_bounds
    
    if i > 0 and start_idx - end_idx_previous < sample_rate * 0.1:
        continue
    
# Extract the segment audio
    segment_audio = audio_data[start_idx:end_idx]
    print(f"Playing Segment {i+1}: Duration {len(segment_audio)/sample_rate} seconds")
    display(Audio(segment_audio, rate=sample_rate))
    end_idx_previous = end_idx
```
